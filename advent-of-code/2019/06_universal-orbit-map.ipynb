# starting from the root

spoiler alert: this was way too complicated. included for posterity.

```python
orbits = defaultdict(list)
bodies = list()

for s in raw:
    left = s[:3]
    right = s[4:]
    orbits[left].append(right)
    bodies.append(left)
    bodies.append(right)
    
bodies = set(bodies)
```

```python
def grow(path):
    children = orbits[path[-1]]
    if len(children) == 1:
        path.append(children[0])
        return grow(path)
    elif len(children) == 0:
        return path
    else:
        new_children = []
        for child in children:
            new_children.append(grow([child]))
        path.append(new_children)
        return path
```

```python
tree = grow(['COM'])
tree
```

```python
def superlen(tree):
    count = 0
    for branch in tree:
        if type(branch) == list:
            count += superlen(branch)
        else:
            count +=1
    return count
```

```python
print(len(bodies))
print(superlen(tree))
```

```python
def supersum(tree):
    total = 0
    for branch in tree:
        if type(branch) == list:
            total += supersum(branch)
        else:
            total += branch
    return total
```

```python
def measure_orbit(branch, base=0):
    measures = []
    indices = np.array([branch.index(b) for b in branch]) + 1
    mask = np.array([(type(b) != list) for b in branch])
    max_index = np.max(indices * mask) + base
    for leaf in branch:
        if type(leaf) == list:
            measures.append(measure_orbit(leaf, max_index))
        else:
            measure = branch.index(leaf) + base
            measures.append(measure)
    return measures
```

```python
print(superlen(measure_orbit(tree)))
print(supersum(measure_orbit(tree)))
```

# starting from the satellite instead of the root

```python
orbits = defaultdict(list)
rev_orbits = {}
bodies = list()

for s in raw:
    left = s[:3]
    right = s[4:]
    orbits[left].append(right)
    rev_orbits[right] = left
    bodies.append(left)
    bodies.append(right)
    
bodies = set(bodies)
```

```python
rev_orbits
```

```python
def measure_orbit(body):
    measure = 0
    while body != 'COM':
        measure += 1
        body = rev_orbits[body]
    return measure
```

```python
measure_orbit('SAN')
```

```python
measures = [measure_orbit(b) for b in bodies]
sum(measures)
```

```python
def get_path(body):
    path = []
    while body != 'COM':
        path.append(body)
        body = rev_orbits[body]
    return path + ['COM']
```

```python
get_path('SAN')
```

```python
get_path('YOU')
```

```python
san = [b for b in get_path('SAN') if b not in get_path('YOU')]
you = [b for b in get_path('YOU') if b not in get_path('SAN')]
# these lists include 'SAN' and 'YOU' but do not include where the orbits are shared
len(san) + len(you) - 2
```
